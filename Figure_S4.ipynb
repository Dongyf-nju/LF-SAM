```python
# -----------------------------导入python库------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import netCDF4 as nc
import sys
# 将目标模块所在的路径添加到 sys.path 中
sys.path.append(r"/home/yfdong/data/work/LF-SAM/code/Library")
from PreprocessVar import (
    get_letter,
    get_letter
)

```

```python
# 读取数据
Positive_df = pd.read_excel(f'/home/yfdong/data/work/SMmerge/CN/merge/code/analysis_code/Revision/EventDateDf_full_Positive.xlsx')
Negative_df = pd.read_excel(f'/home/yfdong/data/work/SMmerge/CN/merge/code/analysis_code/Revision/EventDateDf_full_Negative.xlsx')
```

```python
Positive_df['CMORPH'] = Positive_df['Observed (mm/day)']/24* Positive_df['Duration (hours)']
Negative_df['CMORPH'] = Negative_df['Observed (mm/day)']/24* Negative_df['Duration (hours)']
Positive_df['WRF-S'] = Positive_df['WRF-S (mm/day)']/24* Positive_df['Duration (hours)']
Positive_df['WRF-H'] = Positive_df['WRF-H (mm/day)']/24* Positive_df['Duration (hours)']        
Negative_df['WRF-S'] = Negative_df['WRF-S (mm/day)']/24* Negative_df['Duration (hours)']
Negative_df['WRF-H'] = Negative_df['WRF-H (mm/day)']/24* Negative_df['Duration (hours)']
```

```python
def DrawBar(ax, ColumnIndex, cols, subTitle, dataframe, BarYrange, IF_Xticks=True):
    # 计算每列的中位数，并将其填充到 NaN 值中
    fontsize = 8.5
    bar_labels = list(cols.keys())
    MEANdf = dataframe[bar_labels]
    MEANdf = MEANdf.sum(numeric_only=True)
    counts = MEANdf.tolist()
    # 选择需要绘制的变量
    color_list = [sub_dict['color'] for sub_dict in cols.values()]
    bars = ax.bar(bar_labels, counts, label=bar_labels, color=color_list, edgecolor=color_list, alpha=0.6, linewidth=1)
    bars = ax.bar(bar_labels, counts, label=bar_labels, color='none', edgecolor=color_list, alpha=1, linewidth=1)
    # 添加数据标签
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height+0.005),
                    xytext=(0, 0),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=fontsize-1,
                    style='italic',
                    
                    bbox=dict(boxstyle='round,pad=0.2',  # 背景框样式：圆角，内边距 0.5
          fc='white',  # 背景颜色：黄色
          alpha=0.5,
          ec='none') # 背景透明度：0.5
                    )
    ax.set_ylim(BarYrange)
    # 主要刻度线设置 - 较长的刻度线
    ax.tick_params(
        axis="both",
        which="major",
        direction="in",
        width=0.5,      # 线宽
        length=3,       # 主要刻度线长度
        color='#363636',
        labelsize=fontsize-1
    )
    ax.grid(True, which="major", linestyle="--", color="gray", linewidth=0.75, alpha=0.3)

    ax.text(0.01, 1.1, f"({get_letter(ColumnIndex)}) {subTitle}", transform=ax.transAxes, 
        fontsize=fontsize-0.5, 
        # fontweight='bold',
        va='top', ha='left')
    pass
AWBcols3 = {
    'CMORPH': {'color': '#1C1C1C', 'linestyle': '-'},
    'WRF-S': {'color': "#FF7256", 'linestyle': '-'},
    'WRF-H': {'color': "#00BFFF", 'linestyle': '--'},
}
IF_Xticks=True
plt.rcParams['axes.edgecolor'] = 'black'  # 边框颜色
plt.rcParams['axes.linewidth'] = 0.5     # 边框线宽
LenVar = 3
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(1*LenVar*1.2, 1.5*1.2), dpi=350,
                        # sharex = True,
                        sharey=True,
                        )
BarYrange = [0, 800]
DrawBar(axs[0], 0, AWBcols3, 'Positive Events', Positive_df, BarYrange, IF_Xticks=IF_Xticks)
DrawBar(axs[1], 1, AWBcols3, 'Negative Events', Negative_df, BarYrange, IF_Xticks=IF_Xticks)
axs[0].set_ylabel('(mm)', fontsize=9)
plt.subplots_adjust(wspace=0.1, hspace=0.1)
```
